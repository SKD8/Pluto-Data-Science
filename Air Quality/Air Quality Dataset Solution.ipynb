# Data Description

## Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

## Importing the DataSet

```python
aq=pd.read_csv('airquality.csv')
aq.head()
```

# Basic EDA

# a) Month and Wind

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20,8
```

```python
sns.boxplot(x = 'Month', y = 'Wind', data = aq)
plt.grid();
```

# b) Day and Wind

```python
sns.boxplot(x = 'Day', y = 'Wind', data = aq)
plt.grid();
```

# c) Day and Solar Radiation

```python
sns.boxplot(x = 'Day', y = 'Solar.R', data = aq)
plt.grid();
```

# d) Day and Mean Ozone

```python
sns.boxplot(x = 'Day', y = 'Ozone', data = aq)
plt.grid();
```

# e) Month and Mean Ozone

```python
sns.boxplot(x = 'Month', y = 'Ozone', data = aq)
plt.grid();
```

## Ozone Levels on 7th July

```python
aq[(aq.Month==7) & (aq.Day==7)]
```

The Ozone levels for july 7th is 77.

## Average temperature in the month of September

```python
aq.groupby('Month')['Temp'].mean().iloc[4]
```

The average temperature in the month of september is 76.9.

## Maximum windspeed observed in July

```python
aq.groupby('Month')['Wind'].max().iloc[2]
```

The maximum windspeed observed in july is 14.9.

## Month which has the day with minimum amount of Solar Radiation

```python
aq.sort_values(by='Solar.R',ascending=True).head(1)
```

The month which has the day with minimum amount of solar radiation is July.

## Month with the coldest day

```python
aq.sort_values(by='Temp',ascending=True).head(1)
```

The month with coldest day is May.

## Hottest day

```python
aq.sort_values(by='Temp',ascending=False).head(1)
```

The month with hottest day is August.

## Day with Ozone Level of 78 and Temperature of 86
